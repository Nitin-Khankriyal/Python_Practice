```python
#Valid Sudukooo............................................
import numpy as np
a=np.array(eval(input("Enter the Array")))
temp=0
k=0
p=0
for i in range(0,9):
    b=a[i]
    if b.sum()!=45:
        print("Not the valid Sudukoo....")
        temp=1
        break
    c=a[:,i]
    if c.sum()!=45:
        print("Not the valid Sudukoo....")
        temp=1
        break
    if i!=0 and i%3==0:
        k=k+3
        p=0
    if i!=0 and i%3!=0:
        p=p+3
    d=a[k:k+3,p:p+3]
    if d.sum()!=45:
        print("Not the valid sudukoo....")
        temp=1
        break
if temp!=1:
    print("It is the valid Sudukoo....")
```

```python
import numpy as np
# Columns: [Age, Math Marks, Science Marks],
data = np.array([[18, 85, 78],   # Student 1
                 [19, 92, 88],   # Student 2
                 [17, 76, 95],   # Student 3
                 [18, 65, 70],   # Student 4
                 [20, 90, 85]    # Student 5
                ])
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
p=np.mean(data[:,0])
p
```

```python
#Extract Math marks of all students.
data[:,1]
```

```python
#Find the highest Science mark.
np.max(data[:,2])
```

```python
#Get details of the student who scored more than 90 in Math.
data[data[:,1]>90]
```

```python
#Increase Math marks of all students by 5.
data[:,1]=data[:,1]+5
data
```

```python
#Find how many students are younger than 19.
data[data[:,0]<19,0].size
```

```python
#Calculate the average marks in each subject (column-wise mean).
np.mean(data[:,1:],axis=0)
```

```python
#Get data of students who scored at least 80 in both subjects.
data[np.sum(data[:,1:],axis=1)>=80]
```

```python
#Replace all Science marks < 75 with 0.
data[data[:,2]<75,2]=0
data
```
